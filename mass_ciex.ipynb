```python
import pandas as pd
```

```python
c = 1
```

```python
def college_info_extractor(r_string):
  rD = pd.read_html(r_string)
  r = dict(rD[0][5])
  if len(rD) == 2:
    raw = dict(rD[1]['Name'])
    myData = []
    for eC in raw.values():
      myData.append(eC)
    global c
    snippet = """<tr>
                             <th scope="row">{}</th>
                             <td>{}</td>
                             <td>{}</td>
                             <td>""".format(c,r[0].title(), r[1].title()) 
    myCode = "<li>{}</li>"
    exData = []
    exData.append("<ul>")
    for eC in myData:
      exData.append(myCode.format(eC))
    exData.append("</ul>")
    print(snippet)
    for e_course in exData:
      print(e_course) 
    print(   """</td></tr>""")
  else:
    print("DataFrame Invalid")
```

```python
uLINK = pd.read_html("https://raw.githubusercontent.com/manideep-malyala/marx/main/au%20ugpg.html")
pData = list(dict(uLINK[0]['Code']).values())
for eCollege in pData:
  college_info_extractor("https://apsche.ap.gov.in/unv_filter2.php?a="+str(eCollege))
  c = c+1
```
